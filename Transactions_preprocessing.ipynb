```python
Transactions_df['brand'].fillna('N/A', inplace = True)
Transactions_df['product_line'].fillna('N/A', inplace = True)
Transactions_df['product_class'].fillna('N/A', inplace = True)
Transactions_df['product_size'].fillna('N/A', inplace = True)
```

```python
Transactions_df_imputed = Transactions_df.copy()
Transactions_df_imputed.head(1)
```

```python
from sklearn.impute import KNNImputer
from sklearn.preprocessing import OneHotEncoder

# Drop unnecessary columns
temp_df = Transactions_df_imputed.drop(['transaction_id', 'customer_id', 'transaction_date'], axis=1)

# Separate numerical and categorical columns
numerical_cols = temp_df.select_dtypes(include='number').columns
categorical_cols = temp_df.select_dtypes(include='object').columns

# One-hot encode categorical columns (even though no missing values, this will be used in KNNImputer)
onehot_encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded_categorical = pd.DataFrame(onehot_encoder.fit_transform(temp_df[categorical_cols]),
                                   columns=onehot_encoder.get_feature_names_out(categorical_cols),
                                   index=temp_df.index)  # Make sure indices match the original dataframe

# Concatenate numerical columns with the one-hot encoded categorical columns
combined_data = pd.concat([temp_df[numerical_cols], encoded_categorical], axis=1)

# Impute missing values using KNNImputer on the combined data
imputer = KNNImputer(n_neighbors=3)
imputed_data = pd.DataFrame(imputer.fit_transform(combined_data),
                            columns=combined_data.columns,
                            index=combined_data.index)

# Extract the imputed numerical columns
imputed_numerical = imputed_data[numerical_cols]

# Update the original DataFrame with the imputed numerical values
Transactions_df_imputed[numerical_cols] = imputed_numerical

# Preview the updated DataFrame
Transactions_df_imputed.head()
```

### **Create unique_product_id**

```python
Transactions_df_imputed['brand_num'] = Transactions_df_imputed['brand'].replace({'Solex':'1', 'Trek Bicycles':'2', 'OHM Cycles':'3', 'Norco Bicycles':'4', 'Giant Bicycles':'5', 'WeareA2B':'6', 'N/A':'0'})
Transactions_df_imputed['product_line_num'] = Transactions_df_imputed['product_line'].replace({'Standard':'1', 'Road':'2', 'Mountain':'3', 'Touring':'4', 'N/A':'0'})
Transactions_df_imputed['product_class_num'] = Transactions_df_imputed['product_class'].replace({'low':'1', 'medium':'2', 'high':'3', 'N/A':'0'})
Transactions_df_imputed['product_size_num'] = Transactions_df_imputed['product_size'].replace({'small':'1', 'medium':'2', 'large':'3', 'N/A':'0'})
```

```python
Transactions_df_imputed['model'] = Transactions_df_imputed['product_first_sold_date'].dt.year
Transactions_df_imputed['model'].fillna(0000, inplace=True)
Transactions_df_imputed['model'] = Transactions_df_imputed['model'].astype(int)
Transactions_df_imputed['model'] = Transactions_df_imputed['model'].astype(str)
```

```python
Transactions_df_imputed['product_id_filled'] = Transactions_df_imputed['product_id'].astype(str).str.zfill(3)
Transactions_df_imputed.head()
```

```python
Transactions_df_imputed['unique_product_id'] =  'P' + Transactions_df_imputed['product_id_filled'] + Transactions_df_imputed['brand_num'] +\
                                                Transactions_df_imputed['product_line_num'] + Transactions_df_imputed['product_class_num'] +\
                                                Transactions_df_imputed['product_size_num'] + Transactions_df_imputed['model']
```
